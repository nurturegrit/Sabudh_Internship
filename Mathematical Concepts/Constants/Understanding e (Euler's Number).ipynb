## Learning 'e' Euler's number

#### Definition - a mathematical constant approximately equal to 2.71828, irrational number, meaning it cannot be exactly expressed as a fraction, and its decimal expansion goes on forever without repeating.

### Significance:
* e arises naturally in the context of growth processes, such as population growth, radioactive decay, and interest calculations in finance.
* It's the base of the natural logarithm, often denoted as ln‚Å°. (We will discuss this later.)

### Mathematical Definition:

One way to define \( e \) is through the limit:
$$
e = \lim_{n \to \infty} \left(1 + \frac{1}{n}\right)^n
$$
This definition shows how \( e \) emerges from compound interest calculations.

```python
import numpy as np
e = np.array([(1 + 1/n) ** n for n in range(1, 10000)])
e[-5:]
```

See how when $${n \to \infty}$$ $${e \to 2.71828}$$

You check this by increasing the value of x in range(1, x) in above statement.

To better understand \( e \), you can consider the process of compounding interest more frequently. Imagine you have $1 and an annual interest rate of 100%:

- If you compound the interest once a year, you get $2.
- If you compound the interest twice a year, you get $$\left(1 + \frac{1}{2}\right)^2 = 2.25$$
- If you compound the interest four times a year, you get $$\left(1 + \frac{1}{4}\right)^4 \approx 2.4414
$$

As you increase the frequency of compounding, the amount approaches $$ e \approx 2.71828 $$.


```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10,7) )
money = [1, 2]
interest = 1.00 # 100 %

# Frequencies of Compounding Interest per year
compounded_per_year = np.arange(1, 11)

# Numpy Vectorization is more efficient
for m in money:
    final_value = (m + m*interest/compounded_per_year) ** compounded_per_year
    ax.plot(compounded_per_year, final_value, label=f"Initial Value: {m} (Final Value: {final_value[-1]:.2f}");

e_value = np.e  # This is the value of e from numpy
ax.axhline(y=e_value, color='r', linestyle='--', label='e ‚âà 2.71828')
#ax.set_yscale("log")
ax.set(title=f"Effect of Initial Value and Compounding Freq on Final Value after compounding {len(compounded_per_year)} times.",
       xlabel="Compounding Frequency per year",
       ylabel="Final Value")
ax.legend()
ax.set_ylim(0, 25)
```

When the Initial Value is 1:

- **Formula**: 
  $$
  \text{final\_value} = \left(1 + \frac{1}{n}\right)^n
  $$
- As \( n \) (the number of compounding periods per year) approaches infinity, the final value approaches \( e \), which is approximately 2.71828. This demonstrates the mathematical definition of \( e \).

When the Initial Value is 2:

- **Formula**: 
  $$
  \text{final\_value} = \left(2 + \frac{2}{n}\right)^n
  $$
- As \( n \) approaches infinity, this can be rewritten as:
  $$
  \text{final\_value} = 2^n \left(1 + \frac{1}{n}\right)^n
  $$
  Here, \(\left(1 + \frac{1}{n}\right)^n\) approaches \( e \) as \( n \to \infty \). So,
  $$
  \text{final\_value} \approx 2^n \cdot e
  $$
- Since \( 2^n \) grows exponentially, the final value will approach infinity as \( n \) increases.


### Why is $ e $ Useful?

- **Exponential Growth and Decay**:
  - $e$ is crucial in modeling situations where things grow or decay at rates proportional to their current value.
  - **Example**: Population growth can be modeled by the function
    $
    P(t) = P_0 e^{rt}
    $
    where $ P_0 $ is the initial population, $ r $ is the growth rate, and \( t \) is time.

- **Compound Interest**:
  - In finance, $ e $ helps calculate continuous compound interest.
  - If you invest an amount $ P $ at an annual interest rate $ r $, compounded continuously, the amount of money after $ t $ years is given by
    $
    A = P e^{rt}
    $

- **Calculus and Differential Equations**:
  - $ e $ is fundamental in calculus, particularly in solving differential equations.
  - The function $ e^x $ is unique because its derivative is itself:
    $
    \frac{d}{dx} e^x = e^x
    $
  - It simplifies many calculations involving growth rates and natural logarithms.

- **Probability and Statistics**:
  - In probability theory, $ e $ appears in the context of distributions, such as the exponential distribution and the normal distribution.
  - It helps in understanding the behavior of random variables and in defining the concept of entropy in information theory.


Let‚Äôs break down what it means for $e$ to be the base of the natural logarithm, and we'll touch on the concept of the natural logarithm itself.

### Understanding the Natural Logarithm

1. **Logarithms in General**:
   - A logarithm answers the question: "To what power must we raise a base number to get a certain value?"
   - For example, in the common logarithm (base 10), $ \log_{10}(100) = 2 $ because $10^2 = 100$.

2. **Natural Logarithm**:
   - The natural logarithm is a specific type of logarithm where the base is $e$. It's denoted as $ \ln(x) $.
   - So, $ \ln(x) $ asks: "To what power must we raise $e$ to get $x$?" that is $\log_{e}(x)$

### The Connection with $e$:

1. **Definition of Natural Logarithm**:
   - The natural logarithm of a number $x$ is defined as the exponent to which $e$ must be raised to obtain $x$.
   - Mathematically: If $ y = \ln(x) $, then $ e^y = x $.

2. **Examples**:
   - $ \ln(e) = 1 $ because $ e^1 = e $.
   - $ \ln(e^2) = 2 $ because $ e^2 = e^2 $.
   - $ \ln(1) = 0 $ because $ e^0 = 1 $.

### Why is the Natural Logarithm Important?

1. **Inverse Relationship**:
   - The natural logarithm and the exponential function are inverses of each other.
   - If you apply the natural logarithm to an exponential function, you get back to the exponent: $ \ln(e^x) = x $.
   - Conversely, if you exponentiate the natural logarithm, you get back the original value: $ e^{\ln(x)} = x $.

2. **Applications in Calculus**:
   - In calculus, $e^x$ and $\ln(x)$ have unique and useful properties. For example, the derivative of $e^x$ is $e^x$, and the derivative of $\ln(x)$ is $ \frac{1}{x} $.

3. **Growth and Decay**:
   - Natural logarithms are commonly used in models involving exponential growth or decay, such as population growth, radioactive decay, and continuous compound interest.

### Visualizing the Relationship:

If you plot the function $y = e^x$ and its inverse $y = \ln(x)$, you‚Äôll see that they are reflections of each other across the line \(y = x\).

- For $y = e^x $, as $x$ increases, $y$ grows very quickly.
- For $ y = \ln(x) $, as $x$ increases, $y$ grows more slowly, and the function is defined only for $x > 0$.
  

```python
import matplotlib.pyplot as plt
import numpy as np

# Making Figure
fig = plt.figure(figsize=(12, 10))
ax0 = fig.add_subplot(221)
ax1 = fig.add_subplot(222)
ax2 = fig.add_subplot(212)

# Getting Our Data
x = np.arange(1, 11)
exp_x = np.exp(x)
ln_x = np.log(x)

fig.suptitle(f"Data on first {len(x)} Integers", fontsize=16, fontweight="bold")

# Plotting Data
ax0.plot(x, exp_x, label="Exponential of x")
ax1.plot(x, ln_x, label="Natural Log of x")
ax0.set(title="Exponential Plot", xlabel="Value of x", ylabel="e^(x)")
ax1.set(title="Natural Log Plot", xlabel="Value of x", ylabel="ln(x) or log(e)x")

# Adding the third plot in the second row
ax2.plot(x, np.exp(ln_x), label="y = x")
ax2.set(title="Inverse Relationship between ln(x) and exp(x)", xlabel="x", ylabel="exp(ln(x))")
ax2.set_aspect("equal")
# Adding legends
ax0.legend()
ax1.legend()
ax2.legend()

plt.tight_layout()
plt.subplots_adjust(top=0.90, hspace=0.3)
plt.show()
```
